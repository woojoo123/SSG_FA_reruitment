
* IP의 경우 이런식으로 나와있음
* IP 열에 convert_ip 함수를 적용하여 일관된 소수점 형식으로 변환
* convert_ip함수는 문자열로 주어진 IP를 공백을 기준으로 나누어 전체 이닝과 소수 이닝 부분을 계산하여 소수점 형식으로 변환하거나, 공백이 없으면 그대로 float로 변환

```python
seleced_columns_df_FA_PITCHER_STARTING.dtypes
```

```python
def convert_ip(ip_str):
    if isinstance(ip_str, str):
        if ' ' in ip_str:
            ip_parts = ip_str.split(' ')
            innings = int(ip_parts[0])
            fraction = ip_parts[1].split('/')
            fractional_innings = int(fraction[0]) / int(fraction[1])
            return innings + fractional_innings
        else:
            return float(ip_str)
    return ip_str
```
